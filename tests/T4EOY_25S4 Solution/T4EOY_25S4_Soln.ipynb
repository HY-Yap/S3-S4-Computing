# Task 1.1 – `Player` Class

Implement `Player` class according to the UML class diagram and the attributes / methods specifications.

You will be assessed, but not limited to, based on the standard output of the following function calls:
- `test_1_1()`

<div style="text-align:right; font-weight:bold">[6]</div>

````python
# Your code for Task 1.1
class Player:
    def __init__(self, name, nickname, role):
        self.name = name
        self.nickname = nickname
        self.role = role
        self.k = 0
        self.d = 0
        self.a = 0

    def get_nickname(self):
        return self.nickname
    
    def set_nickname(self, nickname):
        self.nickname = nickname

    def inc_k(self):
        self.k += 1

    def inc_d(self):
        self.d += 1

    def inc_a(self):
        self.a += 1

    def kda(self):
        if self.d > 0:
            d = self.d
        else:
            d = 1
        return round((self.k + self.a) / d, 2)
    
    def __str__(self):
        return f"{self.nickname} ({self.role})" \
            + f" - K: {self.k}, D: {self.d}, A: {self.a}" \
                + f" KDA: {self.kda()}"

    def __lt__(self, other):
        return self.kda() < other.kda()

````

# Task 1.3

Implement a function `quick_sort(p_list)` that takes in a `list` of `Player` objects and sorts them in **descending** order based on their KDA.

You will be assessed, but not limited to, based on the standard output of the following function calls:
- `test_1_3()`

<div style="text-align:right; font-weight:bold">[3]</div>

````python
# Your code for Task 1.3

def quick_sort(p_list):
    if len(p_list) <= 1:
        return p_list
    
    pivot = p_list[0]
    left, right = [], []

    for x in p_list[1:]:
        if pivot < x:
            left.append(x)
        else:
            right.append(x)

    return quick_sort(left) + [pivot] + quick_sort(right)
````

Create the tables in an SQLite database named `esports.db`.

````python
# Your code for Task 2.1
# consider to use "CREATE TABLE IF NOT EXISTS Tablename",
# to avoid error if table already exists

table_players = """
CREATE TABLE IF NOT EXISTS "Players"(
	"PlayerID"	INTEGER,
	"Name"	TEXT NOT NULL,
	"Nickname"	TEXT NOT NULL,
	"Role"	TEXT NOT NULL CHECK("Role" IN ('Top', 'Jungle', 'Mid', 'ADC', 'Support')),
	PRIMARY KEY("PlayerID" AUTOINCREMENT)
);
"""
table_teams = """
CREATE TABLE IF NOT EXISTS "Teams" (
	"TeamID"	INTEGER,
	"Name"	TEXT NOT NULL,
	"School"	TEXT NOT NULL,
	PRIMARY KEY("TeamID" AUTOINCREMENT)
);
"""
table_team_rosters = """
CREATE TABLE IF NOT EXISTS "TeamRosters" (
	"RosterID"	INTEGER,
	"PlayerID"	INTEGER NOT NULL,
	"TeamID"	INTEGER NOT NULL,
	PRIMARY KEY("RosterID" AUTOINCREMENT),
	FOREIGN KEY("PlayerID") REFERENCES "Players"("PlayerID"),
	FOREIGN KEY("TeamID") REFERENCES "Teams"("TeamID")
);
"""
table_games = """
CREATE TABLE IF NOT EXISTS "Games" (
	"GameID"	INTEGER,
	"TeamID1"	INTEGER NOT NULL,
	"TeamID2"	INTEGER NOT NULL,
	"Team1Win"	INTEGER NOT NULL CHECK("Team1Win" IN (0, 1)),
	"Date"	TEXT NOT NULL,
	FOREIGN KEY("TeamID1") REFERENCES "Teams"("TeamID"),
	FOREIGN KEY("TeamID2") REFERENCES "Teams"("TeamID")
);
"""
table_game_stats = """
CREATE TABLE IF NOT EXISTS "GameStats" (
	"StatID"	INTEGER,
	"GameID"	INTEGER NOT NULL,
	"PlayerID"	INTEGER NOT NULL,
	"Kills"	INTEGER NOT NULL DEFAULT 0,
	"Deaths"	INTEGER NOT NULL DEFAULT 0,
	"Assists"	INTEGER NOT NULL DEFAULT 0,
	PRIMARY KEY("StatID" AUTOINCREMENT),
	FOREIGN KEY("GameID") REFERENCES "Game"("GameID"),
	FOREIGN KEY("PlayerID") REFERENCES "Players"("PlayerID")
);
"""
````

````python
# Test code for Task 2.1

# The following code to create the database and tables.
# If your database is locked, try deleting it and re-run the following code to reset
import sqlite3
db_file = 'esports.db'
queries = [
    table_players,
    table_teams,
    table_team_rosters,
    table_games,
    table_game_stats
]
conn = sqlite3.connect(db_file)

for q in queries:
    conn.execute(q)
    
conn.commit()
conn.close()
````

# Task 2.2

The `csv` files inside the `data_files` folder contain information about the players, teams, rosters, games, and game statistics for the first 2 rounds of the tornament.

The first row of each file contains the header of the respective columns. Each row in the files is a comma-separated list of values.

Write a Python program to insert all information from the files into the database.	

<div style="text-align:right; font-weight:bold">[7]</div>


````python
import csv
table_files = {
    'Players': 'data_files/players.csv',
    'Teams': 'data_files/teams.csv',
    'TeamRosters': 'data_files/team_rosters.csv',
    'Games': 'data_files/games.csv',
    'GameStats': 'data_files/game_stats.csv'
}

def read_csv_to_db(table):
    # Your code for Task 2.2
    file_name = table_files[table]
    conn = sqlite3.connect(db_file)
    with open(f'{file_name}', 'r') as f:
        csvreader = csv.reader(f)
        header = next(csvreader)

        placeholders = ','.join(['?' for _ in header])
        
        query = f"""
        INSERT INTO {table} 
        ({','.join(header)}) 
        VALUES 
        ({placeholders})
        """
        
        for row in csvreader:
            try:
                conn.execute(query, tuple(row))
            except sqlite3.Error as e:
                if 'UNIQUE constraint failed' in str(e):
                    continue
                else:
                    print(f'Error inserting row: {row}')
                    print(e)
    conn.commit()
    conn.close()
    
for table in table_files:
    read_csv_to_db(table)
````

````python
def role_ranking(role):
    # Your code for Task 2.3
    conn = sqlite3.connect(db_file)
    query = '''
    SELECT GameID, Teams.Name, Players.Name, Nickname, Kills, Deaths, Assists
    FROM GameStats
    INNER JOIN Players
    ON GameStats.PlayerID = Players.PlayerID
    INNER JOIN TeamRosters
    ON Players.PlayerID = TeamRosters.PlayerID
    INNER JOIN Teams
    ON Teams.TeamID = TeamRosters.TeamID
    WHERE Players.Role = ?
    ORDER BY Kills DESC
    '''

    cursor = conn.execute(query, (role,))
    data = cursor.fetchall()
    cursor.close()
    conn.close()

    return data
````
