````python
# Task 1.1
# Please include your code here.
def read_file(file):
    with open(file, "r") as f:
        # process the first line, save the word as a str
        word = f.readline().strip()
        # print(word)

        # process the rest of the file, save as a 2d list
        lst = []
        line = f.readline()
        while line:
            # save the characters to a lst
            chr_lst = list(line.strip())
            # print(chr_lst)

            # add the line to the main lst
            lst.append(chr_lst)
            line = f.readline()
        
        # return the tuple
        return (word, lst)

# print(read_file("data_files/input.txt"))
````

````python
# Task 1.2
# Please include your code here.
def arr_2d_to_str(lst):
    result = ""
    # iterate through every line
    for line in lst:
        # iterate through every chr
        for chr in line:
            # save chr to result
            result += chr
        # next line
        result += "\n"
    return result[:-1]
````

````python
# Task 1.3
# Please include your code here.
dir = {
        "left": (0, -1, "-"),
        "right": (0, 1, "-"),
        "up": (-1, 0, "|"),
        "down": (1, 0, "|")
}

def locate_first_chr(chr, arr_2d):
    # for the very first chr
    for r in range(len(arr_2d)):
        for c in range(len(arr_2d[0])):
            if arr_2d[r][c] == chr:
                return (r, c)

def locate_chr(prev_coords, chr, arr_2d):
    # when prev chr is known
    # test every direction
    for key in dir:
        x, y = prev_coords[0], prev_coords[1] # reset x an y
        shift_x, shift_y = dir[key][0], dir[key][1]
        x += shift_x
        y += shift_y
        while 0 <= x < len(arr_2d) and 0 <= y < len(arr_2d[0]): # keep checking until out of range
            if arr_2d[x][y] == chr:
                return (key, x, y)
            x += shift_x
            y += shift_y

def connect_word(word, arr_2d):
    prev_coords = locate_first_chr(word[0], arr_2d) # locate the first chr
    for i in range(1, len(word)):
        curr_coords = locate_chr(prev_coords, word[i], arr_2d) # locate the current chr
        # update the arr
        x, y = prev_coords[0], prev_coords[1]
        shift_x, shift_y = dir[curr_coords[0]][0], dir[curr_coords[0]][1]
        # print(x, y)
        # print(word[i], curr_coords[0], (shift_x, shift_y))
        x += shift_x
        y += shift_y
        while arr_2d[x][y] != word[i]:
            # continue until the character is found
            # print(x, y)
            if arr_2d[x][y] == ".":
                arr_2d[x][y] = dir[curr_coords[0]][2]
            else:
                arr_2d[x][y] = "+"
            x += shift_x
            y += shift_y
        # print("found", word[i])
        prev_coords = (curr_coords[1], curr_coords[2])
    return arr_2d_to_str(arr_2d)
````

# Task 3. Get String from Nested Lists

### Description

A word can be hidden inside a set of nested lists.

For example: `[[["T", ["e"]] ,"s"], "t"]` contains the word `"Test"`.

Implement a recursive program `get_word`which takes in a `list` and return the hidden word as a `str`.

### ~ Evidence 3 ~

- Program code for `get_word`.

- You will also be assessed based on the standard output of the following function calls.

    - `print(get_word(lst_0))`

    - `print(get_word(lst_1))`

    - `print(get_word(lst_2))`

#### Marks: \[6\]

````python
# Task 3
# Please include your code here.
def get_word(lst):
    result = ""
    for ele in lst:
        if isinstance(ele, list):
            result += get_word(ele)
        else:
            result += ele
    return result
````
