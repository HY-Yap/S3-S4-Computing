```python
def bubble_sort(lst):
    iteration_count = 0
    for i in range(len(lst) - 1):
        swapped = False
        for j in range(len(lst) - 1 - i):
            if lst[j] > lst[j + 1]:
                # print(f"swapped {lst[j]} and {lst[j + 1]}")
                lst[j], lst[j + 1] = lst[j + 1], lst[j]
                swapped = True
            iteration_count += 1
        if not swapped:
            break

    return iteration_count


def quick_sort(lst):
    if len(lst) <= 1:
        return lst
    pivot = lst[0]
    left = []
    right = []
    for i in range(1, len(lst)):
        if lst[i] < pivot:
            left.append(lst[i])
        else:
            right.append(lst[i])

    return quick_sort(left) + [pivot] + quick_sort(right)


def read_file():
    book_lib_dict = {}  
    with open("input.txt", "r") as f:
        line = f.readline()
        while line:
            if line == "\n":
                line = f.readline()
                continue

            split_list = line.split(" ")
            lib = " ".join(split_list[:-1])
            count = int(split_list[-1])

            books = f.readline().split(",")
            print(books)

            for book in books:
                book = book.strip()
                if book in book_lib_dict:
                    book_lib_dict[book].add(lib)
                else:
                    book_lib_dict[book] = set([lib])

            line = f.readline()
    return book_lib_dict


def output_sorted(book_lib_dict):
    output_file = "results.txt"
    result = ""

    # sort animals by name using bubble sort
    books = list(book_lib_dict.keys())
    bubble_sort(books)

    for book in books:
        # animal are the keys
        result += "<{}>:{}\n".format(book, len(book_lib_dict[book]))

        # sort countries by name using quick sort
        libs = list(book_lib_dict[book])
        sorted_libs = quick_sort(libs)
        for lib in sorted_libs:
            result += lib + ";"
        result += "\n"

    with open(output_file, "w") as f:
        f.write(result)

    print("Done")


data_dict = read_file()
output_sorted(data_dict)


def check_result():
    with open("results.txt") as f1:
        f1_content = f1.read()
    with open("expected_results.txt") as f2:
        f2_content = f2.read()

    return f1_content == f2_content


print(check_result())
```
