```python
# Task 1.1
class AC_Object:
    def __init__(self, name: str, symbol: str):
        self._name = name
        self._symbol = symbol
    
    def get_name(self) -> str:
        return self._name
    
    def __str__(self) -> str:
        return self._symbol

class Plant(AC_Object):
    def __init__(self, name: str, type: str, symbol: str, days_to_mature: int):
        super().__init__(name, symbol)
        self._type = type
        self._days_to_mature = days_to_mature
        self._day_timer = 0
    
    def get_type(self) -> str:
        return self._type
    
    def get_day_timer(self) -> int:
        return self._day_timer
    
    def pass_one_day(self):
        self._day_timer += 1
    
    def check_mature(self) -> bool:
        if self._day_timer >= self._days_to_mature:
            return True
        else:
            return False
    
    def mature(self) -> list:
        if self.check_mature() == False:
            return None
        elif self._type == "flower":
            self._symbol = "¥"
            return None
        elif self._type == "tree":
            fruit_name = self._name.strip(" Tree")
            fruit_1 = AC_Object(fruit_name, "@")
            fruit_2 = AC_Object(fruit_name, "@")
            fruit_3 = AC_Object(fruit_name, "@")
            return [fruit_1, fruit_2, fruit_3]
```

```python
# Task 1.2
class Map:
    def __init__(self, max_row: int, max_col: int):
        self._map = [[None for _ in range(max_col)] for _ in range(max_row)]
        self._max_row = max_row
        self._max_col = max_col
    
    def add_new_obj(self, new_obj: AC_Object, row: int, col: int):
        if self._map[row][col] is not None:
            print(f"Unable to add object. The position is already occupied by {self._map[row][col]}.")
        elif type(new_obj) == Plant and new_obj.get_type() == "tree":
            if (row != 0 and row != self._max_row) and (col != 0 and col != self._max_col):
                for r in range(row-1, row+2):
                    for c in range(col-1, col+2):
                        if self._map[r][c] != None:
                            return
                self._map[row][col] = new_obj
        else:
            self._map[row][col] = new_obj
    
    def pass_a_day(self):
        for r in range(self._max_row):
            for c in range(self._max_col):
                if type(self._map[r][c]) == Plant:
                    self._map[r][c].pass_one_day()
                    if self._map[r][c].check_mature():
                        if self._map[r][c].get_type() == "tree":
                            fruits = self._map[r][c].mature()
                            self._map[r-1][c] = fruits[0]
                            self._map[r][c-1] = fruits[1]
                            self._map[r][c+1] = fruits[2]
                        else:
                            # plant
                            self._map[r][c].mature()
    
    
    def display(self) -> str:
        border = "+" + "-"*self._max_col + "+"
        print(border)
        for r in self._map:
            row = "|"
            for obj in r:
                if obj is None:
                    row += " "
                else:
                    row += str(obj)
            print(f"{row}|")
        print(border)
```

```python
# Test
m = Map(10, 20)

# Add 5 rock objects
m.add_new_obj(AC_Object("Rock", "*"), 0, 0)
m.add_new_obj(AC_Object("Rock", "*"), 2, 3)
m.add_new_obj(AC_Object("Rock", "*"), 0, 18)
m.add_new_obj(AC_Object("Rock", "*"), 8, 12)

# Attempt to add 1 Tree and 2 flowers
m.add_new_obj(Plant("Apple Tree", "tree", "Y", 4), 4, 5)
m.add_new_obj(Plant("Rose", "flower", "v", 3), 4, 5)
m.add_new_obj(Plant("Lily", "flower", "v", 3), 7, 9)


for i in range(5):
    print("Day {}:".format(i))
    m.display()
    m.pass_a_day()
```
