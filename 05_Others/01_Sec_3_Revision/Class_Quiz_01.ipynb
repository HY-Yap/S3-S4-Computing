Write a python function `check_vrn()` to validate if an VRN is valid. Returns `True` if the VRN is valid, `False` otherwise.

You may assume all inputs are in the valid format.

<div style="text-align:right; font-weight:bold">Marks: [5]</div>

````python
# 9, 4, 5, 4, 3, 2
# AZYXUTSRPMLKJHGEDCB

# Your Codes Here
def check_vrn(vrn):
    sum_ = 0
    weight_table = "945432" # length of 6
    cc_table = "AZYXUTSRPMLKJHGEDCB"
    vrn_mid = vrn[1:-1] # length of 6

    # 2. sum of prod
    for i in range(len(vrn_mid)):
        if i < 2: # vrn part is letter
            letter_value = ord(vrn_mid[i]) - ord("A") + 1
            sum_ += letter_value * int(weight_table[i])
        else:
            # vrn part is digit
            sum_ += int(vrn_mid[i]) * int(weight_table[i])
    # print(sum_)

    # 3. remainder
    r = sum_ % 19

    # 4. check code
    check_code = cc_table[r]
    
    return check_code == vrn[-1]

print(check_vrn("SYC3214T"))
````

````python
# Test Cases
print(check_vrn("SYC3214T"))  # True
print(check_vrn("SYC3214B"))  # False

print(check_vrn("SBA1234G"))  # True
print(check_vrn("SBA1234B"))  # False

print(check_vrn("SMO3252L"))  # True
print(check_vrn("SMO3252B"))  # False
````
